```python
import json
import sys
try:
    
        with open("abhi.json","r") as file:
            data=json.load(file)
        #print(data)
        #account_number
        acc_no=int(input("enter account number:"))
        if acc_no!=data["account_number"]:
            print("this account number is not exist ")
            sys.exit()
        #ATM PIN
        AtmPin=int(input("enter atm pin:"))
        if AtmPin!=data["atm_pin"]:
            print("wrong pin")
            sys.exit()
        #ACCOUNT TYPE
        acc_type=input("enter type of account:").lower()
        if acc_type!=data["account_type"]:
            print("not a valid a account type")
            sys.exit()
        #MOBILE NUMBER
        contact=int(input("enter contact:"))
        if contact!=data["mobile number"]:
            print("mobile number is not match")
            sys.exit()
        #BALANCE
        amount=int(input("enter amount:"))
        if amount<=0:
            print("no deduction")
        elif amount>data["balance"]:
            print("not sufficient amount")
        else:
            data["balance"]-=amount
            print("remaining balance",data["balance"])
        with open ("abhi.json",mode="w") as file:
          content=json.dump(data,file,indent=4)
except SystemExit as e:
    print("terminate transaction",e)
except TypeError as e:
    print("please enter valid input type",e)
except ValueError as e:
    print("please enter original value",e)
except Exception as e:
 print(e)
finally:
    print("complete task")

```
